```python
import os
from ase.io import read, write

# Define the source and target directories
Input_index = "_6"
source_dir = f"/nfshome/okresa/Bachelor_Thesis_Office/Project_Vacancy_Intersetiell_ASE_model/input/LAMMPS_input/Active_input{Input_index}/POSCAR_Structures"
target_dir = f"/nfshome/okresa/Bachelor_Thesis_Office/Project_Vacancy_Intersetiell_ASE_model/input/LAMMPS_input/Active_input{Input_index}/mini_data_Structures"

# Check if the target directory exists, if not, create it
if not os.path.exists(target_dir):
    os.makedirs(target_dir)

# Iterate through all .vasp files in the source directory
for filename in os.listdir(source_dir):
    if filename.endswith(".vasp"):
        # Define the full file path for the .vasp file
        source_file_path = os.path.join(source_dir, filename)
        
        # Read the .vasp file using ASE
        atoms = read(source_file_path)
        
        # Wrap the atoms (if needed, this makes sure atoms are within the periodic box)
        atoms.wrap()
        
        # Define the target file path for the .mini.data file
        target_file_name = filename.replace(".vasp", ".mini.data")
        target_file_path = os.path.join(target_dir, target_file_name)
        
        # Write the atoms to the target file in LAMMPS data format
        write(target_file_path, atoms, format="lammps-data", specorder=['Ga', 'Cu', 'Ag', 'Se'])
        
        print(f"Processed and wrote {filename} to {target_file_path}")
```

```python
#Molecular dynamics ver | equal Temp
#WIP implement temp functionality 

import os
import shutil
from pathlib import Path
import datetime

# Source directory where POSCAR files are located
source_directory = '/nfshome/okresa/Bachelor_Thesis_Office/Project_Vacancy_Intersetiell_ASE_model/input/LAMMPS_input/Active_input/mini_data_Structures'

# Destination directory where you want to create the folders
destination_directory = '/nfshome/okresa/Bachelor_Thesis_Office/Project_Vacancy_Intersetiell_ASE_model/input/LAMMPS_input/Active_input/'

# List of additional files to copy (from Active_input)
additional_files = [
    "equilibrate_MD.in",
    "submit-lammps-ml_MD_GPU_Version.sh"
]

# Temperatures to use
Temperatures = [300] 

# Function to modify submit script with temperature
def modify_submit_script(file_path, temperature):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    with open(file_path, 'w') as file:
        for line in lines:
            if '-var temp' in line:
                # Replace the line with the correct temperature
                line = f'~/LAMMPS/Executable_LAMMPS/lmp -k on g $SLURM_NTASKS -sf kk -pk kokkos newton on neigh half -in equilibrate.in -var temp {temperature} | tee out.log\n'
            file.write(line)

# Get a list of .data files
data_files = [filename for filename in os.listdir(source_directory) if filename.endswith('.data')]
print(data_files)

# Loop through each data file
for data_file in data_files:
    data_path = os.path.join(source_directory, data_file)
    print(data_path)

    folder_name = f"0_MD_{data_file}" 
    folder_path = os.path.join(destination_directory, folder_name)

    os.makedirs(folder_path, exist_ok=True)

    for Temperature in Temperatures:
        temp_folder = os.path.join(folder_path, f'T_{Temperature}')
        os.makedirs(temp_folder, exist_ok=True)

        # Copy the data file
        destination_path = os.path.join(temp_folder, 'mini.data')
        shutil.copy(data_path, destination_path)

        for file_name in additional_files:
            src = os.path.join(destination_directory, file_name)
            dst = os.path.join(temp_folder, file_name)
            shutil.copy(src, dst)

            # Modify the temperature in the shell script
            if file_name == "submit-lammps-ml_MD_GPU_Version.sh":
                modify_submit_script(dst, Temperature)

        print(f"Copied and modified {data_file} for T={Temperature} at {temp_folder}")

    print(f"Processed {data_file}")

# Timestamp
now = datetime.datetime.now()
print("\nMD Version\nCopying and script execution completed.")
print("\nTimestamp:")
print(now.strftime("%m-%d-%Y %H:%M"))
```
